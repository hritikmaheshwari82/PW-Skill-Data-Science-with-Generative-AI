# Theory Questions

**Q1. What is NumPy, and why is it widely used in Python?**

NumPy (Numerical Python) is a fundamental Python library used for numerical and scientific computing.

It provides multi-dimensional arrays (ndarray) and many functions to perform efficient operations on them.

Why it's widely used:

Faster and more memory-efficient than regular Python lists.

Supports vectorized operations (no need for explicit loops).

Useful in data analysis, machine learning, and scientific computations.

**Q2. How does broadcasting work in NumPy?**  

Broadcasting allows NumPy to perform arithmetic operations on arrays of different shapes and sizes.

It automatically stretches the smaller array across the larger one without copying data.

#Example

a = np.array([1, 2, 3])        # Shape: (3,)

b = np.array([[10], [20]])     # Shape: (2,1)

result = a + b                 # Shape becomes (2,3) via broadcasting

**Q3. What is a Pandas DataFrame?**

A DataFrame is a 2D labeled data structure with columns of potentially different data types.

Think of it like a table or spreadsheet.

It offers flexible indexing, filtering, grouping, reshaping, and more.

**Q4. Explain the use of the `groupby()` method in Pandas.**  

The groupby() method is used to split data into groups, apply a function (like sum, mean, count), and combine results.

Use cases: aggregation, transformation, filtering of grouped data.

**Q7. What is a heatmap, and when should it be used?**  

A heatmap is a graphical representation of data using colors to show the magnitude of values.

Often used for correlation matrices, frequency counts, and confusion matrices.

Helpful for visualizing patterns, relationships, or concentrations in data.

**Q8. What does the term “vectorized operation” mean in NumPy?**  

A vectorized operation means applying operations element-wise to entire arrays without using loops.

Improves performance and reduces code complexity.

**Q17. Explain the difference between `apply()` and `map()` in Pandas.**

| Works on | Series (1D only)            | DataFrame (rows or columns)        |
| Use case | Element-wise transformation | Row/column-wise custom operations  |
| Example  | `df['col'].map(func)`       | `df.apply(func, axis=1 or axis=0)` |

**Q21. Why is NumPy’s array slicing faster than Python’s list slicing?**

NumPy arrays are contiguous blocks of memory, whereas lists are arrays of pointers.

Slicing in NumPy returns a view (no copy), making it memory-efficient and faster.

Optimized in C at the backend.
